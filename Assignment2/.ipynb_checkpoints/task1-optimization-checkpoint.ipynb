# ECBM E4040 Assignment 2 - Task 1: Optimization

In this task, we introduce multiple SGD-based optimization methods. As you have learned from the last assignment, SGD is an efficient method to update parameters. However, to make SGD perform well, we need to find an appropriate learning rate and a good initial value. Otherwise, the network will get stuck if learning rate is small, or it will diverge if the learning rate is too large. In reality, since we have no prior knowledge about the training data, it is not trivial to find a good learning rate manually. Also, when the network becomes deeper, for each layer we may need to set a different learning rate, and that will again increase the development workload. Obviously, this is not a good direction. 

Another common problem is the lack of sufficient training data. This can make our training get stuck when using the naive SGD method. 

So, **how to set a good learning rate?** You are going to experiment with **SGD with momentum**, **RMSProp**, **Adam**, **Nadam** and make comparisons.

All of these optimizers are adaptive learning rate methods. Here is a useful link: http://ruder.io/optimizing-gradient-descent/.

## Load CIFAR 10

Here we use a small dataset with only 2500 samples to simulate "lack-of-data" situation.

```python
# Load the raw CIFAR-10 data.
X_train, y_train, X_test, y_test = load_data()
X_val = X_train[:500,:]
y_val = y_train[:500]
X_train = X_train[500:2500,:]
y_train = y_train[500:2500]

mean_image = np.mean(X_train, axis=0).astype(np.float32)
X_train = X_train.astype(np.float32) - mean_image
X_val = X_val.astype(np.float32) - mean_image

# We've vectorized the data for you. That is, we flatten the 32×32×3 images into 1×3072 Numpy arrays.
print('Training data shape: ', X_train.shape)
print('Training labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
```

Answer: Overall, the RMS seems to do the best on this dataset, while the SGD does the worst (which is expected). 

SGD is the least robust of these algorithms as it only takes a step by step gradient descent approach with no history factor. Positives: it can jump from minima to other, more optimal minima and is simple to impliment. Negatives: it oscillates about the minima, taking a long time to settle to the exact minima. 

SGD + Momentum fixes many of the issues found in pure SGD. With momentum, the SGD is able to evaluate the history of descent in order to optimize its next step. Positives: the absolute minima is found more quickly (as the momentum is more likely to keep the SGD moving towards the absolute minima rather than getting stuck in a local minima) and it allows the SGD to settle more quickly. Negatives: relies on initial starting point being far enough from a local minima that it gains enough momentum to get out of said local minima.  

RMSprop updates its learning parameters to remove the oscillations found in SGD, much like the SGD + Momentum. This allows for the RMSprop to "jump" local minima towards the absolute minima in the first few steps, and quickly settle into the absolute in later timestamps. It also uses a decay rate to evaluate how important previous time steps were (decaying average of past squared gradients). Positives: doesn't oscillate about minima. Negatives: can get stuck in local minima if initialized in a difficult place

The Adam Optimizer again computes its learning parameter adaptively. It is similar to the RMSprop method, however, it also combines the decay rate with a momentum parameter. This allows for the optimizer to not overshoot as it combines the momentum factor with a decay acting as a damping coefficient. Positives: evaluates history from two perspectives (combining the positives of momentum with RMSprop). Negatives: several hyperparamters to optimize in order to get good results and likes to find "flat" minima rather than steep minima.

The Nadam optimizer is again an update of the Adam optimizer. Nadam takes the Adam optimizer and adds a Nesterov accelerated gradient to optimize more efficiently. By adding Nesterov, the momentum parameter is updated more accurately by evaluating the gradient before the step is taken using a look ahead method. Positives: most robust of all of the algorithms by using a look forward method along with several "history" terms. Negatives: lots of hyperparameters to optimize
